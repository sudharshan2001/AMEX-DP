```python

import numpy as np
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score  
from sklearn.metrics import precision_score                         
from sklearn.metrics import recall_score
from lightgbm import LGBMRegressor,LGBMClassifier
```

```python
train_data = pd.read_parquet('../input/amex-parquet/train_data.parquet')
train_data.drop(['S_2'],axis=1,inplace=True)
train_data=train_data.dropna(axis=1, thresh=int(0.80*len(train_data)))         
train_data.head()
```

```python
# Removing columns with null values above certain threshold
numeri_types = [i for i in train_data.columns if train_data[i].dtype in ['float32','int64']]

to_remove = []
for i in train_data[numeri_types].columns:
    if (train_data[i].isnull().sum()/train_data[i].shape[0]) > 0.03:
        to_remove.append(i)
        train_data = train_data.drop(i, axis=1)
```

```python
train=train_data.groupby('customer_ID').tail(1)
train=train.set_index(['customer_ID'])
```

```python
train_D63 = pd.get_dummies(train[['D_63']])
train = pd.concat([train, train_D63], axis=1)
train = train.drop(['D_63'], axis=1)

train_D64 = pd.get_dummies(train[['D_64']])
train = pd.concat([train, train_D64], axis=1)
train = train.drop(['D_64'], axis=1)
```

```python
del train_D63, train_D64
```

```python
train_without_target=train.drop(['target'],axis=1)
cor_matrix = train_without_target.corr().abs()
upper_tri = cor_matrix.where((np.triu(np.ones(cor_matrix.shape), k=1) + np.tril(np.ones(cor_matrix.shape), k=-1)).astype(bool))

to_drop = [column for column in upper_tri.columns if any(upper_tri[column] > 0.90)]
train_drop_highcorr=train_without_target.drop(to_drop,axis=1)
train_drop_highcorr.shape
```

```python
del train_without_target, cor_matrix, upper_tri
```

```python
from sklearn.feature_selection import VarianceThreshold
from itertools import compress

def fs_variance(df, threshold):

    vt = VarianceThreshold(threshold = threshold)
    fit_vt = vt.fit(df)
    feat_select = list(compress(list(df.columns), vt.get_support()))
    
    return feat_select

columns_to_keep=fs_variance(train_drop_highcorr, 0.06)
train_final=train[columns_to_keep]
len(columns_to_keep)
```

```python
del train_drop_highcorr
```

```python
train_final1=train_final.join(train['target'])
x_train=train_final1.drop(['target'],axis=1)
y_train=train_final1['target']
```

```python
del train_data, train, train_final1,train_final
```

```python
from sklearn.impute import SimpleImputer
imp_mean = SimpleImputer(strategy='mean')
imp_mean.fit(x_train)
imputed = imp_mean.transform(x_train)
```

```python
def amex_metric(y_true, y_pred):
    labels = np.transpose(np.array([y_true, y_pred]))
    labels = labels[labels[:, 1].argsort()[::-1]]
    weights = np.where(labels[:,0]==0, 20, 1)
    cut_vals = labels[np.cumsum(weights) <= int(0.04 * np.sum(weights))]
    top_four = np.sum(cut_vals[:,0]) / np.sum(labels[:,0])
    gini = [0,0]
    for i in [1,0]:
        labels = np.transpose(np.array([y_true, y_pred]))
        labels = labels[labels[:, i].argsort()[::-1]]
        weight = np.where(labels[:,0]==0, 20, 1)
        weight_random = np.cumsum(weight / np.sum(weight))
        total_pos = np.sum(labels[:, 0] *  weight)
        cum_pos_found = np.cumsum(labels[:, 0] * weight)
        lorentz = cum_pos_found / total_pos
        gini[i] = np.sum((lorentz - weight_random) * weight)
    return 0.5 * (gini[1]/gini[0] + top_four)


def lgb_amex_metric(y_pred, y_true):
#     y_true = y_true.get_label()
    return 'amex_metric', amex_metric(y_true, y_pred), True
```

```python
model = LGBMClassifier(boosting_type = 'gbdt', n_estimators = 10000)
```

```python
max_ = np.zeros(x_train.shape[1])

for i in range(2):

    train_features, valid_features, train_y, valid_y = train_test_split(imputed, y_train.values, test_size = 0.2, random_state = i)

    model.fit(train_features, train_y, early_stopping_rounds=300, eval_set = [(valid_features, valid_y)], 
              eval_metric = 'auc', verbose = 1)

    max_ += model.feature_importances_
```

```python

max_ = max_ / 2
max_ = pd.DataFrame({'feature': list(x_train.columns), 'importance': max_}).sort_values('importance', ascending = False)

max_.head()
```

```python
important_features = max_[max_["importance"]>100]['feature'].values
```

```python
x_train=x_train[important_features]
```

```python
from sklearn.impute import SimpleImputer
imp_mean1 = SimpleImputer(strategy='mean')
imp_mean1.fit(x_train)
imputed = imp_mean1.transform(x_train)
```

```python
train_features, valid_features, train_y, valid_y = train_test_split(imputed, y_train.values, test_size = 0.1, random_state = 69)
```

```python
model.fit(train_features, train_y, early_stopping_rounds=200, eval_set = [(valid_features, valid_y)], 
                  eval_metric = 'auc', verbose = 1)
```

```python
del train_features, valid_features, train_y, valid_y , x_train, y_train, imputed
```

```python
test_data = pd.read_parquet('../input/amex-parquet/test_data.parquet',columns=np.insert(important_features,0,"customer_ID"))
```

```python

test_data=test_data.groupby('customer_ID').tail(1)
test=test_data.set_index(['customer_ID'])
```
